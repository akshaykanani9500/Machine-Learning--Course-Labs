```python
iris=data
```

```python
df = pd.DataFrame({'x': iris.data[:,0],'y': iris.data[:,1],'cluster' : iris.target})
df.head()
```

```python
Centroids = {}
k=3
for i in range(k):
    result = list()
    result.append(df.loc[df['cluster'] == i]['x'].mean())
    result.append(df.loc[df['cluster'] == i]['y'].mean())
    
    Centroids[i] = result
  
Centroids
```

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], c=iris.target)
plt.xlabel('Sepa1 Length', fontsize=10)
plt.ylabel('Sepal Width', fontsize=10)
```

```python
class_color = {0: 'g', 1: 'r', 2: 'b'}
for i in range(k):
    plt.scatter(Centroids[i][0],Centroids[i][1], color=class_color[i])
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], c=iris.target,alpha = 0.4)
col = [0,1]
for l in range(k):
  for i in Centroids.keys():
      plt.scatter(Centroids[i][0],Centroids[i][1], c=class_color[i])
plt.show()
```

```python
def distance_finder(df, Centroids):
    for i in range(k):
        df['distance_to_{}'.format(i)] = (np.sqrt((df['x'] - Centroids[i][0]) ** 2+ (df['y'] - Centroids[i][1]) ** 2))
    
    centroid_distance_cols = ['distance_to_{}'.format(i) for i in Centroids.keys()]
    df['nearest_class'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['nearest_class'] = df['nearest_class'].map(lambda x: int(x.lstrip('distance_from_')))
    df['target'] = df['nearest_class'].map(lambda x: class_color[x])
    
    return df
```

```python
df = distance_finder(df, Centroids)
df.head()
```

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], color=df['target'], alpha=0.25)
for i in Centroids.keys():
    plt.scatter(*Centroids[i], color=class_color[i])

plt.show()
```

```python
def centroid_update(Centroids):
    for i in range(k):
        Centroids[i][0] = np.mean(df[df['nearest_class'] == i]['x'])
        Centroids[i][1] = np.mean(df[df['nearest_class'] == i]['y'])
    return Centroids
```

update centroid

```python
Centroids = centroid_update(Centroids)
Centroids
```

graph with updated centroid

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], color=df['target'], alpha=0.25)
for i in Centroids.keys():
    plt.scatter(*Centroids[i], color=class_color[i], edgecolor='k')

plt.show()
```

```python
df = distance_finder(df, Centroids)
```

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], color=df['target'], alpha=0.5)
for i in Centroids.keys():
    plt.scatter(Centroids[i][0],Centroids[i][1], color=class_color[i], edgecolor='k')
plt.show()
```

update untill we got the same centroid

```python
while True:
    nearest_Centroids = df['nearest_class'].copy(deep=True)
    Centroids = centroid_update(Centroids)
    df = distance_finder(df, Centroids)
    if nearest_Centroids.equals(df['nearest_class']):
        break
```

```python
Centroids = centroid_update(Centroids)
Centroids
```

```python
fig = plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], color=df['target'])
for i in Centroids.keys():
    plt.scatter(Centroids[i][0],Centroids[i][1], color=class_color[i], edgecolor='k')

plt.show()
```
